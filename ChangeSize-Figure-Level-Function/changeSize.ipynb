# Change size in figure level function

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data set

tips = pd.read_csv('/Usdatasets/tips.csv')
tips
```

### To change the size of figure level function we user height and aspect

```python
sns.displot(data=tips,x='tip',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```
