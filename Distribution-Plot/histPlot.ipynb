# Histogram plot ( hist plot)

```python
import numpy as np
import pandas as pd
import plotly.express as px # using this only for dataset not for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data set

tips = pd.read_csv('datasets/tips.csv')
tips
```

### Axis level function -> histplot()

```python
sns.histplot(data=tips,x='total_bill')
```

### Figure level function --> displot()

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

### bins parameter

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
```

## It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
## Discrete bins are automatically set for categorical variables

```python
# count plot
sns.displot(data=tips,x='day',kind='hist')
```

### hue parameter

```python
sns.displot(data=tips,x='tip',kind='hist',bins=20,hue='sex')
```

### element parameter

```python
sns.displot(data=tips,x='tip',kind='hist',bins=20,element='step')
```
