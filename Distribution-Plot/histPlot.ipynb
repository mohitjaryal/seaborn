# Histogram plot ( hist plot)

```python
import numpy as np
import pandas as pd
import plotly.express as px # using this only for dataset not for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data set

tips = pd.read_csv('/Users/mohit/Documents/seaborn/datasets/tips.csv')
tips
```

### Axis level function -> histplot()

```python
sns.histplot(data=tips,x='total_bill')
```

### Figure level function --> displot()

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```
