# Kdeplot -> (it's an axis level function)
## kde plot-> Kernel Density Estimate plot

```python
import numpy as np
import pandas as pd
import plotly.express as px # using this only for dataset not for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = pd.read_csv('/Users/mohit/Documents/seaborn/datasets/tips.csv')
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

### using figure level function -> displot()

## hue & fill parameter 

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```
