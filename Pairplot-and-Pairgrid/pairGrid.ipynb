## Plotting Pairwise Relationship (PairGrid Vs Pairplot)

### Pair Grid

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('datasets/iris.csv')
iris
```

```python
# PairGrid
g = sns.PairGrid(data=iris)
# g.map()
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris,hue='species')

# we can set any graph we want to plot like doxplot, violinplot etc  
g.map_diag(sns.histplot) # diginally -> all hsitogram
g.map_offdiag(sns.histplot) # off digonally -> all scatterplot
```

### kdeplot and histplot

```python
g = sns.PairGrid(data=iris,hue='species')

g.map_diag(sns.histplot) # diginally -> all hsitogram
g.map_offdiag(sns.kdeplot) # off digonally -> all scatterplot
```

## map_diag -> map_upper -> map_lower

```python
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### vars

```python
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
