# Heatmap
# Plot rectangular data as a color-encoded matrix

```python
import numpy as np
import pandas as pd
import plotly.express as px # using this only for dataset not for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loadind data set
gap = px.data.gapminder()
temp = gap.pivot(index='country',columns='year',values='lifeExp')
temp
```

### There's only 1 way of plotting heatmap -> .heatmap(), it is an axis level function and there's no figure level function

```python
# heatmap 
plt.figure(figsize=(15,10))
sns.heatmap(temp)
```

```python
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
temp_df
```

#### annot

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True,linewidth=0.5)
```
