## Facet Plot
### This only work with figure level function
#### Means it will only work with relplot()
#### It will now work with scatterplot() and lineplot() -> because they are axis level function

```python
import numpy as np
import pandas as pd
import plotly.express as px # using this only for dataset not for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loadind data set

tips = pd.read_csv('/Users/mohit/Documents/seaborn/datasets/tips.csv')
tips
```

## displaying the data (male vs female)

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex')
```

## Using col & row parameter -> makes multiple plot for particular categorical column

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='smoker')
```

## style parameter -> when we want to know extra layer of information

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time')
```

### col_wrap() parameter if we have multiple col --> it controls how the column facets are wrapped into multiple rows

```python
# loadind data set
gap = px.data.gapminder()
gap

# plotting graph
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```
