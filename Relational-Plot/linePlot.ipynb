## Line plot 

```python
import numpy as np
import pandas as pd
import plotly.express as px # using this only for dataset not for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

## Setting global styles

```python
sns.set_style("whitegrid")
plt.style.use( 'dark_background')
```

```python
# loadind data set -> using plotly.express 
gap = px.data.gapminder()

# filtering the data
ind = gap[gap['country'] == 'India']
ind
```

### line plot -> it is an axis level function

```python
sns.lineplot(data=ind,x='year',y='lifeExp')
```

## Plotting this smae graph -> using figure level function
### .relplot().  -> relation plot

```python
sns.relplot(data=ind,x='year',y='lifeExp',kind='line')
```

## Line plot for 3 countries

```python
# filtering data to get the data of only 3 countries
temp = gap[gap['country'].isin(['India','Brazil','France'])]
temp
```

```python
# relplot() -> with hue parameter
sns.relplot(kind='line', data=temp,x='year',y='lifeExp',hue='country')
```

## Adding continent information using style() parameter

## style parameter -> when we want to know extra layer of information

```python
sns.relplot(data=temp,x='year',y='lifeExp',kind='line',hue='country',style='continent')
```

## size parameter -> It maps a numeric or categorical variable to the size (area) of each point in your scatter plot.

#### here size means thickness & thinenss of the line

```python
sns.relplot(data=temp,x='year',y='lifeExp',kind='line',hue='country',style='continent',size='continent')
```
