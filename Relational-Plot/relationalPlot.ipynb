## Started learning Seaborn 

```python
import numpy as np
import pandas as pd
import plotly.express as px # using this only for dataset not for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loadind data set

tips = pd.read_csv('/Users/mohit/Documents/seaborn/datasets/tips.csv')
tips
```

### scatter plot -> it is an axis level function

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

## Plotting this smae graph -> using figure level function
### .relplot().  -> relation plot

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

## Using hue parameter

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex')
```

## style parameter -> when we want to know extra layer of information

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time')
```

## size parameter -> It maps a numeric or categorical variable to the size (area) of each point in your scatter plot.

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```
