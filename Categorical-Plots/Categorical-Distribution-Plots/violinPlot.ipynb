# Violinplot = (Boxplot + KDEplot)

### violin plot 

```python
import numpy as np
import pandas as pd
import plotly.express as px # using this only for dataset not for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = pd.read_csv('/Users/mohit/Documents/seaborn/datasets/tips.csv')
tips
```

### voilin plot -> axis level 

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

### catplot -> figure level function

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

### hue parameter

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
```

### if you want to make side by side plot -> use split

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
```
