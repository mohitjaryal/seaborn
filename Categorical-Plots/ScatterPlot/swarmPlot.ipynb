```python
import numpy as np
import pandas as pd
import plotly.express as px # using this only for dataset not for plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

#### swarm plot -> it is and axis level function

```python
sns.swarmplot(data=tips,x='day',y='total_bill') # use jitter=False -> if u want data in same line
```
